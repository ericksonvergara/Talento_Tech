```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_excel(r'C:\\Talento_Tech\\rendimiento_estudiantes.xlsx')
print(datos)
```

```python
print(datos.describe())
```

```python
print(datos.isnull().sum())
```

```python
datos = datos.dropna()
print(datos.isnull().sum())
```

```python
print(datos.duplicated().sum())
```

```python
datos = datos.drop_duplicates()
print("\nNúmero de registros duplicados después de eliminarlos:")
print(datos.duplicated().sum())
```

```python
print(datos)
```

```python
datos.to_excel("datos_limpios.xlsx",index=False)
```

```python
#Transformar datos
rendimiento = []
for calificacion in datos["Calificacion"]:
    if calificacion >= 80:
        rendimiento.append("Alto")
    else:
        rendimiento.append("Bajo")
print('El rendimiento es: ', rendimiento)
```

```python
datos["rendimiento"] = rendimiento
print(datos)
```

```python
#Graficas con matplotlb
scatter_matrix = pd.plotting.scatter_matrix(datos, figsize=(5, 5))
plt.show()

plt.scatter(datos['Horas_Estudio'],datos['Calificacion'])
plt.title('Horas de estudio vs Calificacion')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificacion')
plt.grid(True)
plt.show()
```

```python
#graficas con seaborn
sns.countplot(x='rendimiento', data=datos)
plt.title('Rendimiento de los estudiantes')
plt.xlabel('Rendimiento')
plt.ylabel('Cantidad')
plt.show()
```
